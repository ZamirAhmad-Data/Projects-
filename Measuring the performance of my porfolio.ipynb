```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import yfinance as yf
plt.style.use('seaborn')
import seaborn as sns
```

```python
ticker = ['TSLA','MSFT','AAPL','BABA','GOOG','AMZN']
```

```python
stocks = yf.download(ticker,start='2015-01-01',end='2022-11-22')
```

```python
stocks.to_csv('tech.csv')
```

```python
stocks= pd.read_csv('tech.csv', header=[0,1],parse_dates=[0],index_col=[0])['Close']
stocks
```

```python
stocks
```

```python
norm= stocks.div(stocks.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize=(10,8),fontsize=15)
plt.title('Normalized Return (2015-Now)')
plt.ylabel('Return %',fontsize=15)
plt.xlabel('Date',fontsize=15)
plt.show()
```

Among all stocks in my portfolio, Tesla performed the best. I earned a return of over 1000% within seven years. Amazon was my second-best stock pick giving me a return of over 600%. My worst-performing stock was Alibaba, with a negative return of about 25%.

```python
norm.tail()
```

```python
ret= stocks.pct_change().dropna()
```

```python
ret.describe().T.loc[:,['mean','std']]
```

```python
summary = ret.describe().T.loc[:,['mean','std']]
```

```python
summary['mean']= summary['mean']*252
summary['std'] = summary['std']*np.sqrt(252)
```

```python
summary
```

```python
summary.plot.scatter(x='std',y='mean',figsize=(12,8),s=50,fontsize=15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,'std']+0.002,summary.loc[i,'mean']+0.002),size=15)
plt.xlabel('ann.Risk(std)',fontsize=15)
plt.ylabel('ann.Return',fontsize=15)
plt.title('Risk/Return',fontsize=15)
plt.show()
```


In this graph, we look at the Risk/Return tradeoff of my stock picks. Tesla, as expected, has the highest standard deviation and return. Amazon has the second-highest return but lower risk. Alibaba has the lowest return but the second highest standard deviation making it a bad purchase relative to my other stocks.
